### [Problem 6] Creating a diagram

```python
import matplotlib.pyplot as plt 
s_setosa =df[df['Species']==0]
s_virgiclor =df[df['Species']==1]
s_virginica =df[df['Species']==2]

samples= [len(s_setosa),len(s_virgiclor),len(s_virginica)]
labels = 'setosa', 'virgiclor',  'virginica'
plt.title('a pie chart of the number of samples for each label')
plt.pie(samples,labels= labels, autopct='%1.1f%%')
plt.show()
```

```python
import seaborn as sb 
fig, axes = plt.subplots(1,2)

sb.boxplot(x="Species",y="sepal_length", data=df, ax=axes[0])
sb.violinplot(x="Species",y="sepal_length", data=df, ax=axes[1])

axes[0].set_xticklabels(['setosa', 'virgiclor', 'viginica'])
axes[0].set_title('sepal_lenghth violinplot')
fig.set_size_inches(14, 5)
plt.show()
```
