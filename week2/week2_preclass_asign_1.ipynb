<h4> [Problem 1] Number of wheat on a 2x2 chess board </h4>

```python
import numpy as np
n_squares = 2*2
small_board_list = np.array([1])
for _ in range(n_squares - 1):
    small_board_list =np.append(small_board_list, 2*small_board_list[-1])
print ( " Arrange wheat on a 2x2 board (list): {} " .format (small_board_list))
```

<h4> [Problem 2] Expansion to n Ã— m mass </h4>

```python
import numpy as np
def arraylist(width, height):
    n_squares = width * height
    small_board_list = np.array([1])
    for _ in range(n_squares - 1):
        small_board_list =np.append(small_board_list, 2*small_board_list[-1])
    return small_board_list.astype (np.uint64)
print ( " Arrange wheat on a nxm board (list): {} " .format (arraylist(8, 8)))
```

<h4> [Problem 3] Total number of wheat</h4>

```python
import  matplotlib.pyplot  as plt
import numpy as np

def arraylist(width, height):
    n_squares = width * height
    small_board_list = np.array([1])
    for _ in range(n_squares - 1):
        small_board_list =np.append(small_board_list, 2*small_board_list[-1])
    return small_board_list.astype (np.uint64)
# print(arraylist(2,2))
mean = arraylist(8,8).mean()
print("average: ", mean)
%matplotlib inline
plt.xlabel("column")
plt.ylabel("number")
plt.title("number in each column")
plt.bar(np.arange(1,(8**2)+1), arraylist(8,8))
# plt.axvline(arraylist(8,8).mean(), color='blue', linewidth=2, linestyle='--')
plt.show()
```

<h4> [Problem 5] How many times the second half is the first half </h4>

```python
n_squares = 8*8
indices_of_squares = np.arange(n_squares)
small_board_ndarray = 2**indices_of_squares

def split_list(list):
    half = len(list)//2
    return list[:half].astype (np.uint64), list[half:].astype (np.uint64)
first_half,second_half = split_list(small_board_ndarray)
print("number of times is ï¼š{}".format(sum(second_half) / sum(first_half)))
```
