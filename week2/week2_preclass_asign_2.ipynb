<h4> [Problem 1] Creation of random numbers </h4>

```python
import numpy as np
def random_2d(n=500):
    np.random.seed(0)
    # return (np.random.rand(2,500).round(3))
    mean = (-3, 0)
    cov = [[0.8,1 ], [1, 0.8]]
    x,y = np.random.multivariate_normal(mean, cov, n).T
    return x,y
 
x,y = random_2d(500)

print (x)
print (y)
    
 
```

<h4> [Problem 2] Visualization by scatter plot </h4>

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(0)
N = 500
x,y=random_2d(N)
area = (10 * np.random.rand(N))**2  # 0 to 15 point radii
plt.title("scatter graph")
plt.scatter(x, y, s=area, color='blue', alpha=0.5)
plt.show()
```

<h4> [Problem 4] Addition of data </h4>


```python
import numpy as np
import matplotlib.pyplot as plt


def random_2d_second(n=500):
    np.random.seed(0)
    # return (np.random.rand(2,500).round(3))
    mean = (0,-3)
    cov = [[1, 0.8],[0.8,1 ]]
    x,y = np.random.multivariate_normal(mean, cov, n).T
    return x,y

np.random.seed(0)
N = 500
x,y=random_2d(N)
xx,yy = random_2d_second(500)
area = (8 * np.random.rand(N))**2  # 0 to 15 point radii
plt.title("scatter graph")
plt.scatter(x, y, s=area, color='blue', alpha=0.5, label="0")
plt.scatter(xx, yy, s=area, color='red', alpha=0.4, label="1")
plt.legend()
plt.show()
```

<h4>[Problem 5] Data combination </h4>

```python
import numpy as np
import matplotlib.pyplot as plt
# print(np.vstack((random_2d(), random_2d_second())))
comb = np.concatenate((random_2d(),random_2d_second()),axis = 1)
plt.title("scatter graph")
plt.scatter(comb[0], comb[1], s=area, color='purple', alpha=0.5)
plt.legend()
plt.show()
```
