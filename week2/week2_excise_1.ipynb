#### [Problem 1] Linear function</h4>

```python
import numpy as np
x_arry =np.arange(-50, 50.1,0.1)
y_arry = 0.5*x_arry + 1
x_arry,y_arry
# x_arry.shape,y_arry.shape
    
```

#### [Problem 2] Array binding


```python
xy_arry = np.stack((x_arry,y_arry),-1)
xy_arry,xy_arry.shape
```

#### [Problem 3] Find the gradient

```python
dfx= np.diff(x_arry)
dfy= np.diff(y_arry)
slope= dfy/dfx
slope,slope.shape
```

#### [Problem 4] Draw a graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear function Graph")
plt.plot(x_arry, y_arry, color="blue", linestyle= "dotted", linewidth=3, markersize=8)
plt.show()
```

```python
plt.xlabel("x")
plt.ylabel("slope")
plt.title("Linear Function")

plt.plot(x_arry[1:],slope,color='green', linestyle= "dotted", linewidth=3, markersize=8)
plt.show()
```

#### [Problem 5] Python functionalization

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
 arry_x = np.arange(*x_range)
 arry_y = function(arry_x)
 arry_xy = np.stack((arry_x, arry_y),-1)
 gradient = np.diff(arry_y)/np.diff(arry_x)
 return arry_xy, gradient
def function1(array_x):
    array_y = 0.5*array_x+1
    return array_y

def function2(array_x):
    array_y = array_x**2
    return array_y
def function3(array_x):
    array_y = 2*array_x**2 + 2**array_x
    return array_y
def function4(array_x):
    array_y = np.sin(array_x**0.5)
    return array_y

# array_xy, gradient
```

```python
array_xy1, gradient1 = compute_gradient(function1)

plt.xlabel("x")
plt.ylabel("Gradient")
plt.title("Linear Function y= 0.5*array_x+1")
plt.plot(array_xy1[:,0],array_xy1[:,1], color='blue', linestyle= "dotted", linewidth=3, markersize=8, label="linear function")
plt.plot(array_xy1[:-1,0],gradient1, color='purple', linestyle= "dashed", linewidth=3, markersize=8, label="gradient")
plt.legend()
plt.show()
```

```python
array_xy2, gradient2 = compute_gradient(function2)

plt.xlabel("x")
plt.ylabel("Gradient")
plt.title("Linear Function y= array_x**2")
plt.plot(array_xy2[:,0],array_xy2[:,1], color='blue', linestyle= "dotted", linewidth=3, markersize=8, label="linear function")
plt.plot(array_xy2[:-1,0],gradient2, color='purple', linestyle= "dashed", linewidth=3, markersize=8, label="gradient")
plt.legend()
plt.show()
```

```python
array_xy3, gradient3 = compute_gradient(function3)

plt.xlabel("x")
plt.ylabel("Gradient")
plt.title("Linear Function y= 2*array_x**2 + 2**array_x")
plt.plot(array_xy3[:,0],array_xy3[:,1], color='blue', linestyle= "dotted", linewidth=5, markersize=8, label="linear function")
plt.plot(array_xy3[:-1,0],gradient3, color='purple', linestyle= "dashed", linewidth=3, markersize=8, label="gradient")
plt.legend()
plt.show()
```

```python
array_xy4, gradient4 = compute_gradient(function4,x_range=(0,50.1,0.1))

plt.xlabel("x")
plt.ylabel("Gradient")
plt.title("Linear Function y= sin(array_x**0.5)")
plt.plot(array_xy4[:,0],array_xy4[:,1], color='blue', linestyle= "dotted", linewidth=3, markersize=8, label="linear function")
plt.plot(array_xy4[:-1,0],gradient4, color='purple', linestyle= "dashed", linewidth=3, markersize=8, label="gradient")
plt.legend()
plt.show()
```

#### [Problem 6] Find the minimum value

```python
def compute_gradient2(function, x_range=(-50, 50.1, 0.1)):
 arry_x = np.arange(*x_range)
 arry_y = function(arry_x)
 min_y_vl = np.min(arry_y)
 min_y_arg = np.argmin(arry_y)
#  arry_xy = np.stack((arry_x, arry_y),-1)
#  gradient = np.diff(arry_y)/np.diff(arry_x)
 return f'the Minimum value of y for this function is {min_y_vl} and its index is {min_y_arg}'
```
